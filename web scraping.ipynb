```python
import numpy as np
import re
from collections import Counter

def calculate_tf(term_frequency):
    max_frequency = np.max(term_frequency)
    tf = term_frequency / max_frequency
    return tf

def calculate_idf(term_occurrences, num_documents):
    idf = np.log((num_documents + 1) / (term_occurrences + 1)) + 1
    return idf

def calculate_tfidf(term_frequency_matrix):
    num_documents = term_frequency_matrix.shape[0]
    term_occurrences = np.count_nonzero(term_frequency_matrix, axis=0)

    tf = calculate_tf(term_frequency_matrix)
    idf = calculate_idf(term_occurrences, num_documents)

    tfidf = tf * idf
    return tfidf

# Read the cleaned text file
file_path = "cleaned_data.txt"
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# Preprocess the text to obtain the term frequency matrix
words = re.findall(r'\w+', text.lower())
word_count = Counter(words)
term_frequency_matrix = np.array(list(word_count.values()))

# Calculate the TF-IDF values
tfidf = calculate_tfidf(term_frequency_matrix)

print(tfidf)
```

```python
def cosine_similarity(vector1, vector2):
    dot_product = np.dot(vector1, vector2)
    norm1 = np.linalg.norm(vector1)
    norm2 = np.linalg.norm(vector2)
    similarity = dot_product / (norm1 * norm2)
    return similarity
similarity = cosine_similarity(tfidf[0], tfidf[1])

print(f"Cosine Similarity: {similarity}")
```

```python
import re
from collections import Counter
import math

def calculate_cosine_similarity(doc1, doc2):
    # Convert documents to lowercase and split into words
    words1 = re.findall(r'\w+', doc1.lower())
    words2 = re.findall(r'\w+', doc2.lower())
    
    # Calculate term frequency (TF) for each word in both documents
    tf1 = Counter(words1)
    tf2 = Counter(words2)
    
    # Calculate document frequency (DF) for each unique word in both documents
    unique_words = set(words1).union(set(words2))
    df = {word: 0 for word in unique_words}
    for word in unique_words:
        if word in words1:
            df[word] += 1
        if word in words2:
            df[word] += 1
    
    # Calculate TF-IDF for each word in both documents
    tfidf1 = {word: (tf * math.log(2 / (1 + df[word]))) for word, tf in tf1.items()}
    tfidf2 = {word: (tf * math.log(2 / (1 + df[word]))) for word, tf in tf2.items()}
    
    # Calculate dot product of TF-IDF vectors
    dot_product = sum(tfidf1[word] * tfidf2[word] for word in set(tfidf1) & set(tfidf2))
    
    # Calculate magnitude of TF-IDF vectors
    magnitude1 = math.sqrt(sum(tfidf1[word]**2 for word in tfidf1))
    magnitude2 = math.sqrt(sum(tfidf2[word]**2 for word in tfidf2))
    
    # Calculate cosine similarity
    cosine_sim = dot_product / (magnitude1 * magnitude2)
    
    return cosine_sim

# Read the cleaned text file
file_path = "cleaned_data.txt"
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# Calculate cosine similarity
cosine_sim = calculate_cosine_similarity(text, text)
print(f"Cosine Similarity: {cosine_sim}")
import numpy as np

cosine_sim = calculate_cosine_similarity(text, text)
print(f"Cosine Similarity: {cosine_sim}")

# Convert the cosine similarity into a NumPy array
cosine_sim = np.array([cosine_sim])
print(cosine_sim.shape)

print(cosine_sim.shape)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np

# Load the BERT model and tokenizer
model_name = "asafaya/bert-base-arabic"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Read the cleaned text file
file_path = "cleaned_data.txt"
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# Split the input text into smaller chunks
max_length = 512
chunks = [text[i:i+max_length] for i in range(0, len(text), max_length)]

# Initialize a list to store the word embeddings
word_embeddings = []

# Process each chunk separately
for chunk in chunks:
    # Tokenize and encode the Arabic text
    tokens = tokenizer.encode(chunk, add_special_tokens=True)
    input_ids = torch.tensor(tokens).unsqueeze(0)

    # Pass the input to the BERT model and get the word embeddings
    with torch.no_grad():
        outputs = model(input_ids)
        embeddings = outputs.last_hidden_state

    # Access the word embeddings
    chunk_embeddings = embeddings.squeeze(0)

    # Append the chunk embeddings to the list
    word_embeddings.append(chunk_embeddings)

# Concatenate the word embeddings from all chunks
word_embeddings = torch.cat(word_embeddings, dim=0)

# Convert word embeddings to a NumPy array
word_embeddings_np = word_embeddings.detach().numpy()

# Print the shape of the word embeddings
print("Word Embeddings Shape:", word_embeddings_np.shape)
```
